## Задание 2, вариант 2

Нам дано распределение Лапласа с неизвестной $\mu$ и $b=1$. Для начала заметим, что плотность распределения симметрична относительно $\mu$ (так как $f(x | \mu) = \frac{e^{-|x-\mu|}}{2}$), а значит, $\mu$ - медиана распределения Лапласа.

Тогда сразу хочется оценить $\mu \sim X_{[n/2]}$

Вспомним теорему об асимптотите среднего члена вариационного ряда, сразу подставив $\alpha=\frac{1}{2}$

$$
\sqrt{n} \cdot \frac{X_{[\frac{n}{2}]} - q_{0.5}}{\sqrt{\frac{1}{2}^2}}\cdot p(q_{0.5}) \sim N(0, 1)
$$

Тогда, так как для распределения Лапласа $q_{0.5}=\mu$, $p(q_{0.5})=0.5$:
$$
\sqrt{n} \cdot (X_{[\frac{n}{2}]} - \mu) \sim N(0, 1)
$$

Теперь выводим доверительный интервал через квантили нормального распределения:

$$
P(-z_{1-\alpha/2} < \sqrt{n}(X_{[n/2]} - \mu) < z_{1-\alpha/2}) = 1 - \alpha
$$

$$
P(X_{[n/2]}-\frac{z_{1-\alpha/2}}{\sqrt{n}} < \mu < X_{[n/2]}+\frac{z_{1-\alpha/2}}{\sqrt{n}})
$$

То есть мы получили доверительный интервал:
$$
\boxed{\big[X_{[n/2]}-\frac{z_{1-\alpha/2}}{\sqrt{n}} ; X_{[n/2]}+\frac{z_{1-\alpha/2}}{\sqrt{n}}\big]}
$$

```python
import numpy as np
import scipy.stats as stats

def experiment(size, n):
    stat = 0
    for _ in range(size):
        X = np.random.laplace(loc=2, size=n)
        m = np.median(X)
        z = stats.norm.ppf(0.975)
        left = m - (z / np.sqrt(n))
        right = m + (z / np.sqrt(n))
        if (left < 2 < right):
            stat += 1
    return stat / size, left, right

print("Первый эксперимент")
stat1, left1, right1 = experiment(1000, 25)
print(stat1, "%")
print(f"[{left1:.4f}, {right1:.4f}]")
print("Второй эксперимент")
stat2, left2, right2 = experiment(1000, 10000)
print(stat2, "%")
print(f"[{left2:.4f}, {right2:.4f}]")
```

Здесь замечаем примерно такой же эффект. При увеличении выборки процент попаданий в доверительный интервал приближается к 95%, а доверительный интервал сужается.
