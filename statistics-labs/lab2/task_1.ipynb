#### Задание 1. Вариант 3.
Методом максимального правдоподобия найти оценку параметра $\theta$ биномиального распределения $Bin(m, \theta)$, считая m известным. Эксперимент при $m=4$, $\theta=1/5$

***

Для решения задачи методом максимального правдоподобия нужно составить саму функцию правдоподобия $L(\theta |x)$ (совместное распределение/вероятность выборки данного нам распредлеления)

$X \equiv Bin(m, \theta) \quad \Rightarrow \quad P(X=x_i) = C_m^{x_i}\theta^{x_i}(1-\theta)^{m-x_i}$

$L(\theta |x) = \prod_{i=1}^{i=n}p(x_i|\theta) = \prod_{i=1}^{i=n}P(X=x_i|\theta) \quad \Rightarrow$

$L(\theta |x) = \prod_{i=1}^{i=n}C_m^{x_i}\theta^{x_i}(1-\theta)^{m-x_i}=(\prod_{i=1}^{i=n}C_m^{x_i}) \cdot \theta^{\sum_{i=1}^nx_i} \cdot (1-\theta)^{nm-\sum_{i=1}^nx_i}$

Прологарифмируем эту функцию, поскольку логарифм избавит нас от произведения (будет легче дифференцировать) и при этом, так как логарифм - возрастающая функция, он ещё и не испортит точку максимума $\theta$.

$\ln(L\theta |x) = \sum_{i=1}^nC_m^{x_i} + (\sum_{i=1}^n x_i)\ln(\theta)+(nm-\sum_{i=1}^n x_i)\ln(1-\theta)$

Возьмём у этой огромной формулы производную по $\theta$, чтобы впоследствии найти точку максимума, приравняв полученную производную к 0.

$\ln L(\theta |x)'_\theta = 0 + \frac{\sum_{i=1}^n x_i}{\theta} - \frac{nm-\sum_{i=1}^n x_i}{1-\theta}=0$

Получается, теперь осталось решить это уравнение относительно $\theta$.

$\frac{\sum_{i=1}^n x_i}{\theta} = \frac{nm-\sum_{i=1}^n x_i}{1-\theta}$

$(\sum_{i=1}^n x_i)(1-\theta) = (nm-\sum_{i=1}^n x_i)\theta$

$\sum_{i=1}^n x_i = \theta nm$

Отсюда получаем оценку максимального правдоподобия:

$\boxed{\hat\theta = \frac{\sum_{i=1}^n x_i}{nm} = \frac{\overline X}{m}}$

### Смещение оценки

$E(\hat\theta) = E\left(\frac{\overline X}{m}\right) = \frac{1}{m}E(\overline X) = \frac{1}{m}E\left(\frac{1}{n}\sum_{i=1}^n x_i\right) = \frac{1}{m} \cdot \frac{1}{n} \sum_{i=1}^n E(x_i) = \frac{1}{mn} \sum_{i=1}^n m\theta = \frac{1}{mn} \cdot n \cdot m\theta = \theta$

(Так как по свойствам биномиального распределения $E(x_i)=m\theta$)

$\boxed{оценка \ несмещённая}$

***
### Эффективность
Докажем, что мы получили эффективную оценку. Для этого нужно найти информацию Фишера и применить неравенство Рао-Крамера, которое покажет нам оценку снизу. Если будет достигнуто равенство, значит, что оценка эффективная. Для применения неравенства нужно так же, чтобы оценка была несмещённой, но это мы уже проверили.

Для выборки $X_1, \dots, X_n \sim Bin(m, \theta)$:

$$
i(\theta) = -E(\frac{\partial^2 \ln{p(x_i, \theta)}}{\partial^2 \theta})
$$

$$
\ln p(x, \theta) = \ln C_m^x + x\ln\theta + (m-x)\ln(1-\theta)
$$

Тогда посчитаем сначала первую, а потом вторую производную:
$$
\frac{\partial \ln p(x_i, \theta)}{\partial \theta} = \frac{x}{\theta} - \frac{m-x}{1-\theta}
$$
$$
\frac{\partial^2 \ln p(x_i, \theta)}{\partial \theta^2} = -\frac{x}{\theta^2} - \frac{m-x}{(1-\theta)^2}
$$

$$
i(\theta) = -E\left(-\frac{X_i}{\theta^2} - \frac{m-X_i}{(1-\theta)^2}\right)
$$

Так как $E(X_i) = m\theta$:
$$
i(\theta) = \frac{E(X_i)}{\theta^2} + \frac{m - E(X_i)}{(1-\theta)^2} = \frac{m\theta}{\theta^2} + \frac{m - m\theta}{(1-\theta)^2}
$$

$$
i(\theta) = \frac{m}{\theta} + \frac{m(1-\theta)}{(1-\theta)^2} = \frac{m}{\theta} + \frac{m}{1-\theta} = \frac{m(1-\theta) + m\theta}{\theta(1-\theta)} = \frac{m}{\theta(1-\theta)}
$$

Тогда применим неравенство Рао-Крамера:
$$
Var(\theta) \geq \frac{1}{ni(\theta)}
$$
И при этом наша оценка ММП: $\hat{\theta} = \frac{\overline X}{m}$

Дисперсия оценки:
$$
Var(\hat{\theta}) = Var\left(\frac{\overline X}{m}\right) = \frac{1}{m^2}Var(\overline X) = \frac{1}{m^2} \cdot \frac{m\theta(1-\theta)}{n} = \frac{\theta(1-\theta)}{mn}
$$

То есть мы как раз получили равенство в оценке неравенством Рао-Крамера:
$$ \frac{\theta(1-\theta)}{mn} = \frac{1}{n \cdot \frac{m}{\theta(1-\theta)}}$$

А значит, наша оценка эффективная!
***

#### Дисперсия оценки

$D(\~\theta)=D(\frac{\overline X}{m})=\frac{D(\overline X)}{m^2}=\frac{\sum_{i=1}^n D(x_i)}{n^2m^2}=\frac{nm\theta(1-\theta)}{n^2m^2}=\frac{\theta(1-\theta)}{nm}$

(Так как это биномиальное распределение, то $D(x_i)=m\theta(1-\theta)$)

$\boxed{D(\~\theta)=\frac{\theta(1-\theta)}{mn}}$

### MSE (Среднеквадратичная ошибка)

$MSE(\~\theta)=E((\~\theta-\theta)^2)=D(\~\theta)+(E(\~\theta)-\theta)^2=D(\~\theta)+0$

$\boxed{MSE(\~\theta)=D(\~\theta)=\frac{\theta(1-\theta)}{mn}}$

### Свойства оценки

Все оценки максимального правдоподобия:
- состоятельные
- асимптотически нормальные
- асимптотически эффективные

А ещё наша оценка получилась несмещённой, но в общем случае для ОМП это бывает не всегда

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import pandas as pd
```

Проведём эксперимент

```python
n = np.array([10, 20, 50, 100, 150, 200, 250, 500, 1000, 2000, 3000, 10000])  # объём выборок
k = 100000 # количество генерируемых выборок для каждого n_i
interval = 0.05 # что-то типа доверительного интервала

m = 4 # количество испытаний в каждом биномиальном эксперименте
t = 1/5 # вероятность успеха в биномиальном распределении

bias_list = []
variance_list = []
mse_list = []
above_interval_list = []
near_true_percent_list = []

for n_i in n:
    theta_list = np.zeros(k)
    for j in range(k):
        generated_data = np.random.binomial(m, p=t, size=n_i)
        theta = np.mean(generated_data) / m #это наша теоретическая оценка из ММП
        theta_list[j] = theta
    difference = theta_list - t

    bias = np.mean(difference) # "выборочное" смещение, которое у нас в теории должно равняться 0
    variance = np.var(theta_list) # выборочная дисперсия
    mse = bias**2 + variance # среднеквадратичная ошибка
    above_interval = np.sum(np.abs(difference) > interval) # количество значений, отклоняющихся более чем на interval
    near_true_percent = np.sum(np.abs(difference) <= interval) / k * 100 # процент значений, попадающих в интервал

    bias_list.append(bias)
    variance_list.append(variance)
    mse_list.append(mse)
    above_interval_list.append(above_interval)
    near_true_percent_list.append(near_true_percent)


    theoretical_variance = t * (1 - t) / (m * n_i)
    theoretical_bias = 0
    theoretical_mse = theoretical_variance


results_df = pd.DataFrame({
    'Объём выборки': n,
    'Смещение (Bias)': np.array(bias_list, dtype=np.float64),
    'Дисперсия': variance_list,
    'Среднеквадратичная ошибка (MSE)': mse_list,
    'Количество значений параметра, отличающегося от истинного значения более чем на 5%': above_interval_list,
    'Процент полученных параметров, попадающих в интервал 5% от истинного значения': near_true_percent_list
})

display(results_df)
```

```python
plt.figure(figsize=(10, 10))

# Зависимость MSE от объема выборки
plt.subplot(2, 2, 1)
plt.plot(n, mse_list, 'r', label='MSE в эксперименте', linewidth=2)
theoretical_mse_values = variance_list
plt.plot(n, theoretical_mse_values, 'b--', label='MSE в теории', linewidth=2)
plt.xlabel('n')
plt.ylabel('MSE')
plt.title('Зависимость MSE от объема выборки')
plt.legend()
plt.grid(True, alpha=0.3)

# Зависимость смещения от объема выборки
plt.subplot(2, 2, 2)
plt.plot(n, bias_list, 'green', label='Смещение в эксперименте', linewidth=2)
plt.axhline(y=0, color='red', linestyle='--', label='Теоретическое смещение = 0')
plt.xscale('log')
plt.xlabel('n')
plt.ylabel('Смещение')
plt.title('Зависимость смещения от объема выборки')
plt.legend()
plt.grid(True, alpha=0.3)

# Зависимость количества больших отклонений от объема выборки
plt.subplot(2, 2, 3)
plt.plot(n, above_interval_list, 'r-', label='Количество |ошибка| > 0.05', linewidth=2)
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Объем выборки n')
plt.ylabel('Количество больших отклонений')
plt.title('Количество выборок с большой ошибкой')
plt.grid(True, alpha=0.3)
```
