### Задание 1

Правда ли, что:

 - Поле *id_purchase* является *primary key* в таблице **orders**? (Проверь уникальность значений, а также наличие нуллов в колонке *id_purchase* в таблице **orders**).
 - Поле *id_client* является *primary key* в таблице **clients**?  (Проверь уникальность значений, а также наличие нуллов в колонке *id_client* в таблице **clients** ).
 - В таблице **orders** в полях *id_client*, *sum_pay*, *id_trading_point*, *date_pay* нет пропусков (нуллов)?
 - В таблице **clients** в полях *gender* и *age* нет пропусков (нуллов)?

Результаты выведи с помощью интерпретируемых (легко читаемых) принтов.

```python
def is_primary_key(column):
    if column.notnull().any() and column.nunique() == len(column):
        print(f"Поле {column.name} является primary key")
    else:
        print(f"Поле {column.name} не является primary key")

def where_are_nulls(columns):
    for col in columns:
        if col.isnull().any():
            print(f"Поле {col.name} содержит пропуски")
        else:
            print(f"Поле {col.name} не содержит пропусков")
print("1) Проверка на primary key")
is_primary_key(df_orders['id_purchase'])
is_primary_key(df_orders['id_client'])
print("2) Проверка на пропуски")
where_are_nulls([df_orders['id_client'], df_orders['sum_pay'], df_orders['id_trading_point'], df_orders['date_pay']])
where_are_nulls([df_clients['gender'], df_clients['age']])
```

Построй в табличной форме динамику трёх показателей по дням:

 - количества заказов,

### Задание 4

Правда ли, что средний чек по женщинам в возрасте от 20 до 30 лет выше среднего чека по мужчинам в этой же возрастной категории?

В ответе выведи, на сколько рублей (0 знаков после запятой) и на сколько процентов (1 знак после запятой) женский чек выше мужского.

```python
merged_df = df_clients.merge(df_orders, on='id_client', how='inner', indicator=True)
merged_df = merged_df[(merged_df['age'] < 30) & (merged_df['age'] >= 20)]
women_avg = merged_df[merged_df['gender'] == 'F']['sum_pay'].mean()
men_avg = merged_df[merged_df['gender'] == 'M']['sum_pay'].mean()
print('Разница в платежах (женщины - мужчины):', round(women_avg - men_avg))
print('Разница в платежах (в процентах):', round((women_avg / men_avg - 1) * 100, 1), "%")
```

### Задание 5

Правда ли, что чем больше возраст, тем больше чек?

Для ответа на вопрос необходимо разбить показатель возраста на интервалы и оценить средний чек в рамках каждого интервала. Попробуй также уловить данную зависимость в рамках каждого пола.

```python
merged_df = df_clients.merge(df_orders, on='id_client', how='inner', indicator=True)
new_df = pd.DataFrame(columns=['age_group', 'women_avg', 'men_avg', 'all_genders_avg'])
for i in range(int(merged_df['age'].min()), int(merged_df['age'].max()) - 4, 5):
    age_group = merged_df[(merged_df['age'] >= i) & (merged_df['age'] < i + 5)]
    women_avg = age_group[age_group['gender'] == 'F']['sum_pay'].mean()
    men_avg = age_group[age_group['gender'] == 'M']['sum_pay'].mean()
    all_genders_avg = age_group['sum_pay'].mean()
    new_df.loc[len(new_df)] = {'age_group': f"{i}-{i+5}", 'women_avg': women_avg, 'men_avg': men_avg, 'all_genders_avg': all_genders_avg}
new_df.reset_index()
new_df
```

### Задание 6

Предположим, что стоимость обработки одной операции в торговой точке (с учётом костов на аренду, продавца и так далее) составляет $250$ рублей.

Для каждой торговой точки рассчитай:
 - количество операций;
 - суммарную выручку;
 - чистую валовую прибыль;
 - валовый средний чек (из расчёта на одну операцию);
 - чистый средний чек (то есть валовый средний чек из расчёта на одну операцию за вычетом костов);
 - уникальное количество покупающих клиентов;
 - валовый средний чек из расчёта на одного клиента.

```python
value = 250
grouped_data = df_orders.groupby('id_trading_point').agg(
    count_of_operations=('id_purchase', 'count'),
    total_sum=('sum_pay', 'sum'),
    average_sum=('sum_pay', 'mean'),
    unique_customers=('id_client', 'nunique')
)

grouped_data['gross_profit'] = grouped_data['total_sum'] - value * grouped_data['count_of_operations']
grouped_data['gross_average'] = grouped_data['gross_profit'] / grouped_data['count_of_operations']
grouped_data['gross_unique'] = grouped_data['gross_profit'] / grouped_data['unique_customers']
grouped_data.head()
```
